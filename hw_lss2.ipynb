```python
np.random.seed(42)
h = np.random.randint(0, 16, 16)
H = h.reshape((4, 4))
D = H * 10 + G
D_det = np.linalg.det(D) #определитель
D_rnk = np.linalg.matrix_rank(D) #ранг матрицы
D_inv = np.linalg.inv(D) # Обратная матрица
D_inv
```

```python
D_inv[D_inv < 0] = 0
D_inv[D_inv > 0] = 1
D_inv
```
